```python
%matplotlib inline
import pandas as pd
import geopandas as gpd
import glob, random
import mio
import shapely
```

```python
def scale(x,y,z):
    return (x/100,y/100,z/100)
```

```python
# check
lfiles = glob.glob('json/*.json')
lgdf = []
for i, file in enumerate(lfiles):
    mio.show_perc(i, len(lfiles), 100)

    gdf = gpd.read_file(file, driver="GeoJSON")
    
    gl = []
    for geom in gdf.geometry:
        geom2 = shapely.ops.transform(scale, geom)
        gl.append(geom2)
        
    gdf.geometry = gl
    lon, lat = gdf.iloc[0].lon, gdf.iloc[0].lat
    x_swiss, y_swiss = mio.wgs_swiss(lon, lat)
    xs, ys = round(x_swiss), round(y_swiss)
    gdf = gdf[gdf.geometry.area>0]
    gdf.geometry = gdf.geometry.translate(xs, ys)
    lgdf.append(gdf) 
```

```python
res = pd.concat(lgdf)
```

```python
res_g = gpd.GeoDataFrame(res)
```

```python
mio.write_tab(res_g, 'test.tab')
```
